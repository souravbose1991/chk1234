```python
import warnings
warnings.filterwarnings("ignore")

from dtm_sb import DynTM2 as dtmSB
import numpy as np
import pandas as pd
import swifter
import re

import nltk
nltk.data.path.append("/app/localstorage/u_am_coe/notebooks/nltk_data")
from nltk.corpus import stopwords

# Plots
import pyLDAvis
import pyLDAvis.gensim  # don't skip this
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
################## Stopwords list ##################

stop_words = set(stopwords.words("english") + ['say','million','year','quarter', 'month', 'thank','company','see', 'hi', 'hello', 'how',
                 'think','get','look','okay','well','question', 'billion','management','be', 'chf', 'usd', 'i', 'I', 'want', 'go', 
                 'thanks','morning','please','afternoon','mean','sort','guess','lot', 'right', 'much', 'would', 'could', 'bit', 
                 'like', 'give','take','expect','wonder','try','might','maybe','may','mr', 'sir', 'ahead', 'lady', 'gentleman'])
```

```python
tran1 = pd.read_csv('ubs_transcripts_extra.csv')
tran2 = pd.read_csv('mytranscripts.csv')
tran = pd.concat([tran1, tran2], axis=0)
tran.reset_index(inplace=True, drop=True)
tran.head()
```

```python
ubs_model_static = dtmSB.DynTM(stop_words=stop_words, num_topics=6, company='UBS')
ubs_model_static.struct_file(tran)
ubs_model_static.subset_trans(631, cutoff=10)
ubs_model_static.subset_df.head()
print(np.asarray((ubs_model_static.uniquetimes, ubs_model_static.time_slices)).T)
```

```python
ubs_model_static.dynamic_model(n_top = 'optimal', timestamp = '2005', seq=False)
```

```python
bny_model_static = dtmSB.DynTM(stop_words=stop_words, num_topics=6, company='BNY')
bny_model_static.struct_file(tran)
bny_model_static.subset_trans(77, cutoff=10)
bny_model_static.subset_df.head()
print(np.asarray((bny_model_static.uniquetimes, bny_model_static.time_slices)).T)
```
